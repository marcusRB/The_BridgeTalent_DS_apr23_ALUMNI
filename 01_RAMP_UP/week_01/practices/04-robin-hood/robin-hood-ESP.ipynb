# Robin Hood: 
Estamos en plena competición para ganar el concurso de tiro con arco en Sherwood. Con nuestro arco y flechas disparamos sobre una diana e intentamos acertar lo más cerca posible del centro. 

El centro de la diana viene representado por los valores (0, 0) en los ejes de coordenadas.

* estructuras de datos: listas, conjuntos, tuplas
* operadores lógicos: if-elif-else

## Descripción: 
En el espacio de 2 dimensiones un punto se puede definir por un par de valores que corresponden a la coordenada horizontal (x) y a la vertical (y). El espacio puede quedar dividido en 4 zonas (cuadrantes): Q1, Q2, Q3, Q4. Cuyo punto de unión único es el punto (0, 0). 

Si un punto se encuentra en Q1 tanto su coordenada x como la y son positivas. Te dejo un enlace a wikipedia para que te familiarices con estos cuadrantes.

## Tareas

```python
puntos = [(4,5), (-0,2), (4,7), (1,-3), (3,-2), (4,5), 
          (3,2), (5,7), (-5,7), (2,2), (-4,5), (0,-2), 
          (-4,7), (-1,3), (-3,2), (-4,-5), (-3,2), 
          (5,7), (5,7), (2,2), (9, 9), (-8, -9)]


print(type(puntos))

# 1. Robin Hood es famoso por acertar a una flecha con otra flecha. ¿Lo ha conseguido? 

#utilizo la funcion len porque elimina los valores repetidos de una lista


lista_sin_duplicados = set(puntos)
lista_sin_duplicados

# si la longitud de la lista puntos es mayor a la lista lista_sin_duplicados, quiere decir que 
# hay valores repetidos que fueron eliminados

if len(puntos) > len(lista_sin_duplicados):
    print("Si, acertó una flecha dentro de otra flecha")
else:
    print("No lo ha conseguido")
 
# 2. Calcula cuántos flechazos han caido en cada cuadrante. 

Q1 = 0
Q2 = 0
Q3 = 0
Q4 = 0

for x,y in puntos:
    if x > 0 and y > 0:
        Q1 += 1
    elif x < 0 and y > 0:
        Q2 += 1
    elif x < 0 and y < 0:
        Q3 += 1
    elif x > 0 and y < 0:
        Q4 += 1
    else: 
        None

print("En el Q1 cayeron: ",Q1,"flechas" )
print("En el Q2 cayeron: ",Q2,"flechas" )
print("En el Q3 cayeron: ",Q3,"flechas" )
print("En el Q4 cayeron: ",Q4,"flechas" )

# 3. Halla el punto más cercano al centro. Calcula su distancia al centro

puntos = [(4,5), (0,2), (4,7), (1,-3), (3,-2), (4,5), 
          (3,2), (5,7), (-5,7), (2,2), (-4,5), (0,-2), 
          (-4,7), (-1,3), (-3,2), (-4,-5), (-3,2), 
          (5,7), (5,7), (2,2), (9, 9), (-8, -9)]


## creamos una funcion que nos devuelva el valor de la distiancia entre las coordenadas dadas
## y la coordenada de origen

def distancia_desde_origen(punto):
    """
    Esta función devuelve la distancia entre el valor de origen y cualquier otra coordenada
    se utiliza el teorema de pitagoras para calcular esta distancia
    
    """
    x, y = punto
    distancia = (x**2 + y**2)**0.5
    return distancia

## probamos la funcion y funciona
punto2 = (puntos[1])
distancia = distancia_desde_origen(punto2)
print(distancia)


## acá no encontraba manera de hacer que me de el minimo, por lo que creé un diccionario, donde
## el valor llave es la tupla y el valor value es la distancia

dic_distancias_desde_origen = {punto: distancia_desde_origen(punto) for punto in puntos}

print(dic_distancias_desde_origen)

## ahora que tengo el diccionario, puedo buscar el valor minimo y devolverlo

distancia_minima = min(dic_distancias_desde_origen.values())

for coordenadas, distancia in dic_distancias_desde_origen.items():
    if distancia == distancia_minima:
        print(coordenadas)


# Definir una función que calcula la distancia al centro puede servir de ayuda. 

## ya la definí

# 4. Si la diana tiene un radio de 9, calcula el número de flechas que hay que recoger al bosque. 

fuera_radio = 0

for coordenadas, distancia in dic_distancias_desde_origen:
    if distancia > 9:
        fuera_radio += 1
        
print(fuera_radio)
```
